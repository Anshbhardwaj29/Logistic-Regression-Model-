```python
import pandas as pd
import matplotlib.pyplot as plt
df  = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
plt.show()
```

```python

plt.figure(figsize=(10,5))
p_df = df[df['Purchased']==1]
np_df = df[df['Purchased']!=1]
plt.scatter(p_df['Age'], p_df['EstimatedSalary'], label = 'Purchased')
plt.scatter(np_df['Age'], np_df['EstimatedSalary'], label = 'Not Purchased')
plt.legend()
plt.show()
```

```python
x = df[['Age', 'EstimatedSalary']].values
y = df['Purchased'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 43)
#Normilzation
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
condusion_matrix = confusion_matrix(y_test, y_pred)
condusion_matrix
```

```python
#Accuracy_score = (TP+TN)/(TP+TN+FN+FP)
accuracy_score = accuracy_score(y_test, y_pred)
accuracy_score
```
